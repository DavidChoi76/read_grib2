```python
rainfall = data["apcpsfc"]
rainfall
```

```python
rain_temporal_loc = rain_loc.loc[dict(time=slice('2023-04-14T03:59:59.999997000', '2023-04-16T23:00:00.000003000'))] 
rain_temporal_loc
```
